**HANDLING NULL VALUES**

```python
data.isnull().sum()
```

```python
data['Age'].fillna(data['Age'].mean(),inplace=True)
```

```python
data.drop('Cabin',axis=1,inplace=True)
```

```python
data['Embarked'].fillna(method='ffill',inplace=True)
```

```python
data.isnull().sum()
```

**HANDLING DUPLICATES VALUES**

```python
data.duplicated().sum()
```

```python
data['Age']=data['Age'].astype('int64')    #changing age datatype object to int
```

```python
data.rename(columns={'Pclass':'passenger_class'},inplace=True)             #renaming the col_name
```

```python
data['passenger_class'] = data['passenger_class'].replace({1: 'First Class', 2: 'Second Class', 3: 'Third Class'})
```

```python
data['passenger_class']=data['passenger_class'].astype('object')    #changing passenger_class datatype int to object
```

```python
data.head()
```

**Exploratory Data Analysis**

```python
survival_counts = data["Survived"].value_counts()
print(survival_counts)
```

```python
plt.figure(figsize = (4,5))
plt.pie(data["Survived"].value_counts(), 
        labels=["not survived", "survived"],
        autopct='%1.1f%%', 
        startangle=90,
       explode=[0, 0.1],
       shadow=True)

plt.title("Total number of Survived People")
plt.axis("equal")
plt.show()
```

```python
data['Sex'].value_counts()
```

```python
# Count the number of males and females
sex_counts = data['Sex'].value_counts()

# Create bar plot
plt.bar(sex_counts.index, sex_counts.values,color=['red','pink'],alpha=0.5)

# Labels and title
plt.xlabel("Sex")
plt.ylabel("Count")
plt.title("Count of Passengers by Gender")
plt.show()
```

```python
x=data.groupby('Sex')['Survived'].sum().index
y=data.groupby('Sex')['Survived'].sum()

plt.figure(figsize=(5,4))
plt.title("Survival Count by Sex on the Titanic")

plt.bar(x,y,edgecolor="black")

plt.xlabel("Sex")
plt.ylabel("Survived")

plt.show()
```

```python
x=data.groupby('Age')['Survived'].sum().index
y=data.groupby('Age')['Survived'].sum()
plt.figure(figsize=(6,4))
plt.title("Survival Rate Across Different Age Groups")

plt.bar(x,y,edgecolor="black")

plt.xlabel("Age")
plt.ylabel("Survived")

plt.show()
```

```python
# number of survived along with the passenger_class
plt.figure(figsize = (6,4))
sns.countplot(data = data, x = data["passenger_class"], hue = data["Survived"])
plt.title("survival rate based on Passenger_class")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Embarked", hue="Survived", data=data)
plt.title("Survival by Embarkation Port")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(data.corr(numeric_only=True), annot=True, fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
#adding col of age_group in dataset
# Define a function to categorize the age
def age_group(age):
    if age <= 12:
        return 'Child'
    elif age <= 19:
        return 'Teen'
    elif age <= 59:
        return 'Adult'
    else:
        return 'Senior'

# Apply the function to create the 'AgeGroup' column
data['AgeGroup'] = data['Age'].apply(age_group)
```

```python
# Plot Survival Rate by Age Group
plt.figure(figsize=(6, 4))
sns.countplot(x='AgeGroup', hue='Survived', data=data)
plt.title("Survival Rate by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Count")
plt.show()
```
